### Question 1

Write a function `setfilter(l,s)` that returns a list containing the items from the list `l` that are in the set `s`.  The items in the returned list must be in the same order as in the list `l`.

For example, `setfilter([1,2,3,4,5],{3,1,5})` should return `[1, 3, 5]`.


```python
def setfilter(l,s):
    a =[]
    for i in l:
        if i in s:
            a.append(i)
    return a

setfilter([1,2,3,4,5],{3,1,5})
```

### Question 2

Write a function `upperdict(l)` that returns a dictionary.  The keys are the items in the list of strings `l` and the values are the keys converted to upper-case.

For example `upperdict(['Apple', 'car', 'aadvark'])` would return `{'Apple': 'APPLE', 'car': 'CAR', 'aadvark': 'AADVARK'}`.

```python
def upperdict(l):
    a = {}
    for i in range(len(l)):
        a[l[i]] = l[i].upper()
    return a
upperdict(['Apple', 'car', 'aadvark'])
```

### Question 3

Write a function `abstan(a,b)` that returns the tangent of the the absolute value of `a` divided by `b` ($\tan(\frac{|a|}{b})$).

For example `abstan(-1,5)` would return `0.2027100355086725`.

```python
import math

def abstan(a,b):
    return math.tan(abs(a)/abs(b))

abstan(-1,5)
```

### Question 4

Write a function `safecalc(a,b)` that raises a ValueError if `a` is greater than `b` or else returns the sum of `a` plus `b`.

For example `safecalc(1,2)` would return `3` while `safecalc(1,2)` would raise a `ValueError` exception.

```python
def safecalc(a,b):
    if a>b:
        raise ValueError
    else:
        return a + b
        
safecalc(1,2)
```

* begin with 0 to 3 upper-case letters, 
* followed by a 5-digit number, and
* end with an optional suffix consisting of a hyphen (-) and a single lower-case letter

For example `serialpat` would match the string "12345-a" or "ABC45678" but not " A98765" or "12345-A".

```python
import re

serialpat = r'^[A-Z]{0,3}\d{5}(-[a-z])?$'

test_strings = ["12345-a", "ABC45678", "A98765", "12345-A"]

for s in test_strings:
    result = "match" if re.match(serialpat, s) else "no match"
    print(f'"{s}" -> {result}')
```
