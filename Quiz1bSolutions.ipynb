### Question 1

Write a function `sumdiff(a,b)` that takes two numbers as arguments and returns a two-element list, the first element of which is `a+b` and the second is `a-b`.

For example, `sumdiff(1,2)` should return the two-element list `[3,-1]`.

```python
def sumdiff(a,b):
    return [a+b, a-b]

sumdiff(1,2)
```

### Question 2

Write a function `offset(l,a)` that takes a list of numbers `l` and an number `a` as arguments and returns the list `l` with `a` added to each element of `l`.

For example, `offset([-1,2,-3],10)` should return `[9, 12, 7, 15]`.

```python
def offset(l,a):
    return [x+a for x in l]

offset([-1,2,-3],10)
```

### Question 3

Write a function `strpair(s1,s2)` where `s1` and `s2` are strings.  The function should returns a list of two-character strings where each string is the concatenation of the letters taken successively from `s1` and `s2`.  You can assume `s1` and `s2` have the same number of characters.

For example `strpair("abz","XYz")` would return `['aX', 'bY', 'zz']`.

```python
def strpair(s1,s2):
    return [a+b for a,b in zip(s1,s2)]
    
strpair("abz","XYz")
```
