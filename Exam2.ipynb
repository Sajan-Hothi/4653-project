### Question 1

Write a function `setge(l,t)` that returns a _set_ containing all the values from the list `l` that are greater than or equal to `t`.

For example, `setge([1,2,3,4,5],3)` should return `{3, 4, 5}`.


````python
def setge2(l,t):
    a = []
    for i in range(len(l)):
        if l[i] >= t:
            a.append(l[i])
    return set(a)

setge2([1,2,3,4,5],3)

def setge(l, t):
    return {x for x in l if x >= t}

setge([1,2,3,4,5], 3)  # {3, 4, 5}
````

### Question 2

Write a function `sqint(l)` that returns a dictionary.  The keys are the integers in the list `l` and the values are the keys squared.

For example `sqint([1,3,9])` would return `{1: 1, 3: 9, 9: 81}`.

````python
def sqint2(l):
    a = {}
    for i in range(len(l)):
        a[l[i]] = l[i]**2
    return a

def sqint(l):
    return {x: x**2 for x in l}

sqint2([1,3,9])
````

### Question 3

Write a function `resist(l)` that returns the harmonic mean of the values in the list `l` as computed by the `statistics.harmonic_mean()` function.

For example `resist([1,2,4,4])` would return `2.0`.

````python
import statistics

def resist(l):
    return statistics.harmonic_mean(l)

resist([1,2,4,4])
````

Define a string `partpat` that is a regular expression that matches the strings that:

* begin with 3-digit number,
* followed by a hypen
* followed by a 3 to 5-digit number, and
* end with an optional suffix consisting of "-ND"

For example `partpat` would match the string "123-456-ND" or "000-11111" but not " 123-456-ND" or "123-56-ND".

````python
import re

partpat = r'^\d{3}-\d{3,5}(-ND)?$'

(re.match(partpat, "123-456-ND" ),
re.match(partpat, "000-11111" ),
re.match(partpat, " 123-456-ND" ),
re.match(partpat, "123-56-ND" ))

# ^  = start    $ = end
re.match(r'^\d$', '5')        # exactly one digit, nothing else

# \d = digit,  \w = word char,  \s = space,  . = anything
re.match(r'\d', '5')          # matches digit
re.match(r'\w', 'a')          # matches letter/digit/_

# {n} = exactly n,  {n,m} = between n and m
re.match(r'\d{3}',   '123')   # exactly 3 digits
re.match(r'\d{3,5}', '1234')  # 3 to 5 digits

# ? = optional,  + = 1 or more,  * = 0 or more
re.match(r'(-ND)?', '-ND')    # -ND is optional
re.match(r'\d+',    '123')    # 1 or more digits

# [A-Z] = range,  [^A] = NOT A,  (abc) = group
re.match(r'[A-Z]{2}', 'AB')   # 2 uppercase letters
re.match(r'[^0-9]',   'a')    # anything not a digit
````
