#### Question 1

Define a function `munge(s,n,r,i)` that is passed a string `s`, an integer `n`, and two floats `r` and `i`.  
The function should return a list containing: a string composed of `s` repeated `n` times and a complex number whose real part is `r` and whose imaginary part is `i`.

For example, `munge('hello',3,2.0,-1)` would return `['hellohellohello', (2-1j)]`.


```python
def munge(s,n,r,i):
    return[s*n, complex(r,i)]
    
munge('hello',3,2.0,-1)
```

#### Question 2

Define a function `is_same(a,b)`
that is passed two objects, `a` and `b`.  The function should return 2 if the objects are the same, 1 if the objects are equal, and 0 otherwise.

For example, after executing the statements `l1=[1]; l2=l1; l3=[1]; l4=[2]` then:
* `is_same(l1,l2)` would return 2,
* `is_same(l1,l3)` would return 1, and
* `is_same(l1,l4)` would return 0.

```python
def is_same(a, b):
    if a is b:
        return 2
    elif a == b:
        return 1
    else:
        return 0

l1 = [1]
l2 = l1
l3 = [1]
l4 = [2]

print(is_same(l1, l2))
print(is_same(l1, l3))
print(is_same(l1, l4))
```

#### Question 3

Declare a function named 
`roundit(a)`
that returns:
* 0.5 if `a` is less than 1, otherwise
* 5 if `a` is less than 10, otherwise
* `a`

For example:
* `roundit(0.1)` would return `0.5`
* `roundit(7)` would return `5`
* `roundit(37)` would return `37`


```python
def roundit(a):
    if a < 1:
        return 0.5
    elif a < 10:
        return 5
    else:
        return a

print(roundit(0.1))
print(roundit(7))
print(roundit(37))
```

#### Question 4

Write a function `longones(l)` that takes a list of string `l` and returns another list 
containing only the strings that have 8 or more characters.

For example, `longones(['a', 'xylophone', 'wombat', 'artichoke'])` would return `['xylophone', 'artichoke']`.

```python
def longones(l):
    m = []
    for i in range(len(l)):
        if len(l[i]) >= 8:
            m.append(l[i])
    return m

print(longones(['a', 'xylophone', 'wombat', 'artichoke']))
```
