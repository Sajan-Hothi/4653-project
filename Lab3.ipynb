#### Question 1

Define a function `uisets(s1,s2)` that takes two string arguments and returns a tuple of two sets.  The first set is the set of all characters in `s1` or `s2`.  The second set is the set of characters in both `s1` and `s2`.

For example, `uisets("aeiou","aardvark")` would return `({'a', 'd', 'e', 'i', 'k', 'o', 'r', 'u', 'v'}, {'a'})`.

```python
def uisets(s1, s2):
    return (set(s1) | set(s2), set(s1) & set(s2))

print(uisets("aeiou", "aardvark"))
```

#### Question 2

Define a function `unshort(s)`
that is passed a string `s` and returns the string with all words shorter than three letters removed.   You can assume words are separated by spaces.

For example, `unshort("The quick brown fox jumps over the lazy dog.")` would return `'quick brown jumps over lazy dog.'`

```python
def unshort(s):
    return ' '.join([w for w in s.split() if len(w) > 3])

print(unshort("The quick brown fox jumps over the lazy dog."))
```

#### Question 3

Declare a string named `partpat` that is a regular expression that matches strings that are exactly the following patterns:

* one to three digits, followed by
* a dash, followed by
* an upper-case letter

OR

* the letter "O", followed by
* exactly three digits

For example, this regular expression would match "25-A", "999-X", and "O001", but would _not_ match " 25-A", "123X", or "O000 ".


```python
import re

partpat = r'^(\d{1,3}-[A-Z]|O\d{3})$'

print(re.fullmatch(partpat, "25-A"))
print(re.fullmatch(partpat, "999-X"))
print(re.fullmatch(partpat, "O001"))
```

#### Question 4

Write a function `utf8(s)` that takes string `s` and returns a byte object that is the UTF-8 encoding of the string.

For example, `utf8("狐狸")` would return `b'\xe7\x8b\x90\xe7\x8b\xb8'`.

```python
def utf8(s):
    return s.encode('utf-8')

print(utf8("狐狸"))
```

#### Question 5

Write a function `sortwords(s)` that takes a string argument `s` and returns a string composed of the words in `s` that are longer than 3 characters, sorted in increasing order and separated by spaces.  Assume the words in `s` are separated by spaces.

For example, `sortwords("the quick brown fox jumps over the lazy dog")` would return `'brown jumps lazy over quick'`.

```python
def sortwords(s):
    return ' '.join(sorted([w for w in s.split() if len(w) > 3]))

print(sortwords("the quick brown fox jumps over the lazy dog"))
```
